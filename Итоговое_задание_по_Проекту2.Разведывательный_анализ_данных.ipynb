## 2. Функции/модули для работы <a name="func"></a>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from scipy.stats import ttest_ind
import math

# для вывода всех столбцов сделаем расширенный формат 
pd.set_option('display.max_rows', 50) # показывать больше строк
pd.set_option('display.max_columns', 50) # показывать больше колонок

def check_count(df_s):
    '''анализ данных в серии: определение количества,
                              соотношение уникальных значений, 
                              количество пустых строк'''
    return display(pd.DataFrame(df_s.value_counts())), display(df_stud.loc[:, [df_s.name]].info()), print(f'Пустых значений: {395 - df_s.count()}({round((395 - df_s.count())/3.95,2)}%)', display(df_stud[df_s.isnull()]))


def blowouts(df_s):
    '''Функция отображения статистической информации по столбцу дата фрейма.
    Принимает столбец датафрейма.
    Выводит основную информацию по данным в столбце, 
    осуществляет проверку на выбросы,
    стоит гистограмму распределения значений'''
    median = df_s.median()
    mean = df_s.mean()
    std = df_s.std()
    IQR = df_s.quantile(0.75) - df_s.quantile(0.25)
    perc25 = df_s.quantile(0.25)
    perc75 = df_s.quantile(0.75)
    l = perc25 - 1.5*IQR
    r = perc75 + 1.5*IQR
    min_df = df_s.min()
    max_df = df_s.max()
    # пустые значения
    col_none = pd.isnull(df_s).sum()
    # количество выбросов
    outliers = df_s.loc[(df_s < perc25 - 1.5*IQR) | (df_s > perc75 + 1.5*IQR)]
    # настройки вида диаграммы
    #df_s.hist(label = df_s.name)
    
    range_hist = ((min_df - ((max_df - min_df)/10)), (max_df + ((max_df - min_df)/10)))
    
    outliers.hist(bins = 20, range = range_hist, color = 'red', label = 'выбросы')
    df_s.loc[df_s.between(perc25 - 1.5*IQR, perc75 + 1.5*IQR)].hist(bins = 20,
                                                                    color = 'green',
                                                                    label = df_s.name)
    if min_df < l or max_df > r:
        return print('Выбросы обнаружены',
                     f'Среднее: {mean}',
                     f'Медиана: {median}',
                     f'Стандартное отклонение: {std}',
                     f'25-й перцентиль: {perc25}',
                     f'75-й перцентиль: {perc75}',
                     f'Интерквартильный интервал(IQR): {IQR}',
                     f'Границы выбросов: {l},{r}',
                     f'Минимальное и максимальное значения: {min_df},{max_df}',
                     f'Количество выбросов:{len(outliers)}',
                     f'Количество пропусков: {col_none}',
                     {display(outliers)},
                     {plt.legend()}, sep='\n')
    else:
        return print('Выбросы не обнаружены',
                     f'Среднее: {mean}',
                     f'Медиана: {median}',
                     f'Стандартное отклонение: {std}',
                     f'25-й перцентиль: {perc25}',
                     f'75-й перцентиль: {perc75}',
                     f'Интерквартильный интервал(IQR): {IQR}',
                     f'Границы выбросов: {l},{r}',
                     f'Минимальное и максимальное значения: {min_df},{max_df}',
                     f'Количество пропусков: {col_none}',
                     {plt.legend()}, sep='\n')

    
def boxplot(col, param):
    '''Функция выводит boxplot график, для визуализации влияния признака переменную.
       Принимает:
       - col - столбец датафрейма (влияющий признак)
       - param - переменная'''
    fig, ax = plt.subplots(figsize = (15, 5))
    sns.boxplot(x=col, y=param, 
                data=df_stud[[col, param]],
                ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('График для ' + col)
    plt.show()
    
def sud_test(data, column, param):
    '''Проверяет, есть ли статистичеки значимые различия в переменной param для колонки column.
       Принимает:
       data - датафрейм
       column - колонка
       param - переменная'''
    cols = data.loc[:, column].value_counts().index
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(data.loc[data.loc[:, column] == comb[0], param], 
                     data.loc[data.loc[:, column] == comb[1], param],
                     nan_policy='omit').pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

## 5.3. Тест Стьюдента по категориальным показателям <a name="ttest"></a>

Тест позволит проверить нулевую гипотезу (изменение в колонке не имеет значимого влияния на изменения в целевых переменных: score и no_score)

```python
#Выделим и посчитаем тудентов с ненулевым баллом
students_nonul_score = df_stud[~pd.isnull(df_stud['score'])]
```

```python
#Для переменной score
for column in df_stud.columns.tolist():
    if column not in (['age', 'absences', 'score', 'no_score']):
        sud_test(students_nonul_score, column, 'score')
```

## 6. Подготовка финальных данных <a name="finalization"></a>

на этом этапе устраним пропуски, выбросы для столбцов, влияющих на целевую переменную

```python
# проходим по каждому из столбцов, найденных ранее

# оказывающие влияние на score: age, absences, address, m_edu, f_edu, m_job, f_job, studytime, 
# failures, schoolsup, goout, 
# оказывающие влияние на no_score: paid, higher, romantic
```

```python
# age - выброс был 1, устранили ранее. Пропуски отсутствуют.
```

```python
# absences - есть 2 значения выбивающихся из общей массы (212, 385) и 12 пропусков, заменяю медианой
df_stud.loc[(df_stud['absences'] > 200.0) | (pd.isnull(df_stud['absences'])), 'absences'] = df_stud['absences'].median()
# проверим
display(blowouts(df_stud.absences))
```

```python
# address - ранее выяснили, что время ы пути жителей сельской местости больше чем у городских.
# по этому можем предусмотреть замену по следующему правилу: 
# если travel_time = 1 или NaN, адрес заполняем U, иначе R
df_stud.loc[pd.isnull(df_stud['address']), 'address'] = df_stud.traveltime.apply(lambda x: 'U' if pd.isnull(x) else 'U' if x==1 else 'R')
# проверим
display(check_count(df_stud.address))
```

```python
# m_edu, f_edu - ранее мы определили, что показатели образования родителей в семъях примерно равны
# посмотрим встречаются ли пропуски сразу в m_edu и f_edu у учеников:
df_stud[(pd.isnull(df_stud['m_edu'])) & (pd.isnull(df_stud['f_edu']))]
```

```python
# таких учеников нет, значит можно заменить пропуски m_edu на значения f_edu и наоборот:
df_stud.loc[pd.isnull(df_stud['f_edu']), 'f_edu'] = df_stud['m_edu']
df_stud.loc[pd.isnull(df_stud['m_edu']), 'm_edu'] = df_stud['f_edu']

# проверим
display(check_count(df_stud.m_edu))
display(check_count(df_stud.f_edu))
```

```python
# m_job и f_job ранее предполагал заменить пропуски наиболее часто врстречающимся значением other, так и поступлю
df_stud.loc[pd.isnull(df_stud['m_job']), 'm_job'] = 'other'
df_stud.loc[pd.isnull(df_stud['f_job']), 'f_job'] = 'other'

# проверим
display(check_count(df_stud.m_job))
display(check_count(df_stud.f_job))
```

```python
# studytime - пропуски заполняем медианой, она еще и с самым частым значением совпадает
df_stud.loc[pd.isnull(df_stud['studytime']), 'studytime'] = df_stud['studytime'].median()

# проверяем
display(blowouts(df_stud.studytime))
```

```python
# failures - пропуски уже заполнили согласно правилу относительно итогового балла 
# 'score' > 55 = 0, 'score' < 55 = 1
```

```python
# schoolsup - пропуски заполним самым частым значением 0
df_stud.loc[pd.isnull(df_stud['schoolsup']), 'schoolsup'] = 0.0

# проверим:
display(check_count(df_stud.schoolsup))
```

```python
# goout - пропуски заполним самым частым значением (оно совпадает с медианой) 3
df_stud.loc[pd.isnull(df_stud['goout']), 'goout'] = 3
# проверим
display(check_count(df_stud.goout))
```

```python
# paid, higher, romantic - оказывают влияние только на no_score, 
# не доконца понятно чем заменить пропуски, пока попробую заменить на самой частой значение
df_stud.loc[pd.isnull(df_stud['paid']), 'paid'] = 0.0
df_stud.loc[pd.isnull(df_stud['higher']), 'higher'] = 1.0
df_stud.loc[pd.isnull(df_stud['romantic']), 'romantic'] = 0.0

# проверим:
display(check_count(df_stud.paid))
display(check_count(df_stud.higher))
display(check_count(df_stud.romantic))
```

<b>Итого для постоения финальной модели у нас есть 2 признака: score и no_score</b>   
И признаки влияющие на них.  

score: age, absences, address, m_edu, f_edu, m_job, f_job, studytime, failures, schoolsup, goout   
no_score: m_edu, failures, guardian, paid, higher, romantic
